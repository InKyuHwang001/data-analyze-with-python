```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('../data/Electric_Production.csv', index_col=0)
data
```

```python
data.index
```

```python
data.index=pd.to_datetime(data.index)
data.index
```

```python
data.head()
```

```python
data.plot(figsize=(20,10),linewidth=5, fontsize=20)
plt.xlabel('years', fontsize=20)
```

```python
#컬럼명 번경
data.columns=['Energy Production']
data.head()
```

```python
value=data[['Energy Production']]
value
```

```python
#이동평균 구하기(단, 기간은 12개월로 함)

value.rolling(12).mean().plot(figsize=(20,10),linewidth=5, fontsize=20)
plt.xlabel('years', fontsize=20)
```

```python
#위 그래프의 차분 그래프 그리기
data.diff().plot(figsize=(20,10),linewidth=5, fontsize=20)
plt.xlabel('years', fontsize=20)
```

```python
# 위 그래프의 acf, pacf 곡선 그리기 
# 사용할 데이터는 data
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(data)
plot_pacf(data)
plt.show
```

```python
diff_1=data.diff(periods=1).iloc[1:]
diff_1.plot()
```

```python
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show
```
