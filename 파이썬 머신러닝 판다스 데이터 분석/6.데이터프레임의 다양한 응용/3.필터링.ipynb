# 불리언 인덱싱

```python
import pandas as pd
import numpy as np
import seaborn as sns

titanic = sns.load_dataset('titanic')
mask1 = (titanic.age >= 10) & (titanic.age < 20) # age 컬럼의 값이 10이상 20미만이면 참 반환
df_teenage = titanic.loc[mask1, :] # mask1의 참인 결과 행만 추출
df_teenage.head(2)
```

```python
mask2 = (titanic.age < 10) & (titanic.sex == 'female') # age 컬럼의 값이 10 미만, sex 컬럼의 값이 female인 경우 참 반환
df_female_under10 = titanic.loc[mask2, :]
df_female_under10.head()
```

```python
mask3 = (titanic.age < 10) | (titanic.age >= 60) # age 컬럼의 값이 10 미만이거나 60이상인 경우 참 반환
df_1060 = titanic.loc[mask3, ['age','sex','alone']] # mask3의 조건에 해당하는 행, age,sex,alone 컬럼만 추출
df_1060.head(3)
```

# isin()

```python
pd.set_option('display.max_columns',10) # 출력 설정 변경, 최대 10개의 컬럼만 출력

mask3 = titanic['sibsp'] == 3
mask4 = titanic['sibsp'] == 4
mask5 = titanic['sibsp'] == 5
df_boolean = titanic[mask3 | mask4 | mask5] # sibsp 컬럼이 3이거나 4이거나 5인 행 추출
df_boolean.head()
```

```python
isin_filter = titanic['sibsp'].isin([3,4,5]) # isin()메소드 사용
df_isin = titanic[isin_filter] # sibsp 컬럼에 3,4,5 중 포함되는 값이 있는 행 추출
df_isin.head()
```
