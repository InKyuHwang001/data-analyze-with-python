## 2.6. Dataframe의 Data 선택, 필터링

- Dataframe[ ] 의 연산자는 Numpy[ ] 와는 다르다.
- Dataframe[ ] 에 입력 가능한 값

 - 컬럼명(또는 컬럼명 리스트)
 - 불린 인덱스
 - 행의 슬라이싱 연산
 
- loc[ , ] : 명칭 기반 인덱싱 연산자

 - 행 : Dataframe의 index (또는 index 리스트)
 - 열 : Dataframe의 컬럼명 (또는 컬럼명 리스트)
  
- iloc[ , ] : 위치 기반 인덱싱 연산자

 - 행 : Dataframe의 행위치
 - 열 : Dataframe의 열위치

## 2.7. Misssing Data

- NaN은 값이 없는 NULL인 경우이다.
- NaN은 sum(), mean() 등의 함수 연산시 제외된다.
- 머신러닝 알고리즘은 NaN 을 처리하지 않는다.
- isna() : NaN 확인
- fillna() : NaN을 다른 값으로 대체
- dropna() : NaN을 제거

```python
df.isna() # na : not available
```

```python
df.isna().sum()
```

```python
df[df.Age.isna()]
```

```python
na = df[df.Age.isna()]
na.sum()
```

```python
df.Embarked.isna().sum() # method chaining 
```

```python
df.Embarked.fillna('N')
```

```python
df.Age.isna().sum()
```

```python
df.Age.dropna()
```

```python
df.dropna()
```
